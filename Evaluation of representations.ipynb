```python
%%R

dir = "25-05-2014/classification"

errors.bagofwords = read.csv(paste(dir, "bagofwords.csv", sep="/"), sep=" ", header = F)[2:4]
errors.bagofwords_binary = read.csv(paste(dir, "bagofwords-binary.csv", sep="/"), sep=" ", header = F)[2:4]
errors.tfidf = read.csv(paste(dir, "tf-idf.csv", sep="/"), sep=" ", header = F)[2:4]
```

```python
%%R

colors <- rainbow(3)
#par(mfrow=c(1,2), mar = c(1,1,0, 2))
plot(c(0,600),c(0.3,0.5), main = "Mutation+Events", type="n", xlab="Minimun size (events)", ylab="Error in classification [0:1]" ) 

legend(400, 0.5, c("bag of words","binary bag of words", "tf-idf"), cex=0.8, col=colors, pch=21:22, lty=1:2)

lines(x=errors.bagofwords[,1], y=errors.bagofwords[,3], type="b", col=colors[1])
lines(x=errors.bagofwords_binary[,1], y=errors.bagofwords_binary[,3], type="b", col=colors[2])
lines(x=errors.tfidf[,1], y=errors.tfidf[,3], type="b", col=colors[3])

plot(c(0,600),c(0.3,0.5), main = "Mutation Only", type="n", xlab="Minimun size (events)", ylab="Error in classification [0:1]" ) 

legend(400, 0.5, c("bag of words","binary bag of words", "tf-idf"), cex=0.8, col=colors, pch=21:22, lty=1:2)

lines(x=errors.bagofwords[,1], y=errors.bagofwords[,2], type="b", col=colors[1])
lines(x=errors.bagofwords_binary[,1], y=errors.bagofwords_binary[,2], type="b", col=colors[2])
lines(x=errors.tfidf[,1], y=errors.tfidf[,2], type="b", col=colors[3])
```
